```python
import plotly.express as px
# Create a bar graph for confirmed cases by country
fig_confirmed = px.bar(
    df,
    x='Country/Region',
    y='Confirmed',
    color='Country/Region',
    title='Confirmed COVID-19 Cases by Country',
    labels={'Confirmed': 'Total Confirmed Cases'}
)

# Show the graph
fig_confirmed.show()
```

```python
# Create a bar graph for confirmed cases by country
fig_confirmed = px.bar(
    df,
    x='Country/Region',
    y='Recovered',
    color='Country/Region',
    title='Recovered COVID-19 Cases by Country',
    labels={'Recovered': 'Total Confirmed Cases'}
)

# Show the graph
fig_confirmed.show()
```

```python
# Create a bar graph for confirmed cases by country
fig_confirmed = px.bar(
    df,
    x='Country/Region',
    y='Deaths',
    color='Country/Region',
    title='Deaths COVID-19 Cases by Country',
    labels={'Deaths': 'Total Death Cases'}
)

# Show the graph
fig_confirmed.show()
```

```python
fig_confirmed = px.histogram(
    df,
    x='Confirmed',
    title='Histogram of Confirmed COVID-19 Cases',
    labels={'Confirmed': 'Total Confirmed Cases'},
    marginal='box'  # Optional: add a box plot for additional insights
)

# Create a histogram for deaths
fig_deaths = px.histogram(
    df,
    x='Deaths',
    title='Histogram of COVID-19 Deaths',
    labels={'Deaths': 'Total Deaths'},
    marginal='box'
)

# Create a histogram for recovered cases
fig_recovered = px.histogram(
    df,
    x='Recovered',
    title='Histogram of Recovered COVID-19 Cases',
    labels={'Recovered': 'Total Recovered Cases'},
    marginal='box'
)

# Show the histograms
fig_confirmed.show()
fig_deaths.show()
fig_recovered.show()
```

```python
selected_columns = ['Confirmed', 'Deaths', 'Recovered']

# Create a DataFrame with selected columns
selected_df = df[selected_columns]

sns.pairplot(selected_df)
plt.suptitle('Pair Plots for COVID-19 Data', y=1.02)
plt.show()
```
